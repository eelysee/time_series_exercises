```python
# this will be the loop to concat the next page into a df
people_root = 'https://swapi.dev/api/people/?page=2'
people = pd.DataFrame()
current = people_root
while True:
    people = pd.concat([people, current], ignore_index=True)
    current = current['next']    
```
